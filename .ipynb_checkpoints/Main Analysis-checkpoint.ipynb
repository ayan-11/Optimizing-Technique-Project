```python
import pandas as pd

# Load the dataset
df = pd.read_csv('restaurant_data_india.csv')
print(df)
```

```python
import pulp

# Create a LP problem
prob = pulp.LpProblem("Maximize_Profit", pulp.LpMaximize)

# Create decision variables
dishes = df['Dish Name']
profit_per_dish = df['Profit per Dish'].values
ingredient_usage = df['Ingredient Usage (units)'].values
prep_time = df['Preparation Time (hours)'].values

# Decision variable for each dish (number of dishes to prepare)
dish_vars = pulp.LpVariable.dicts("Dishes", dishes, lowBound=0, cat='Integer')

# Define the objective function (maximize total profit)
prob += pulp.lpSum([profit_per_dish[i] * dish_vars[dishes[i]] for i in range(len(dishes))]), "Total_Profit"

# Define constraints
# Example constraints (you can modify them based on your project requirements):
# 1. Ingredient availability constraint
available_ingredients = 20  # Total available units of ingredients
prob += pulp.lpSum([ingredient_usage[i] * dish_vars[dishes[i]] for i in range(len(dishes))]) <= available_ingredients, "Ingredient_Availability"

# 2. Chef working hours constraint
available_hours = 10  # Total available chef hours
prob += pulp.lpSum([prep_time[i] * dish_vars[dishes[i]] for i in range(len(dishes))]) <= available_hours, "Chef_Working_Hours"

# Step 3: Solve the LP problem
prob.solve()

# Print the results
print("Status:", pulp.LpStatus[prob.status])
for dish in dishes:
    print(f"{dish}: {dish_vars[dish].varValue}")
print("Total Profit:", pulp.value(prob.objective))
```

```python
import matplotlib.pyplot as plt

# Get the values for visualization
prepared_dishes = [dish_vars[dish].varValue for dish in dishes]
plt.bar(dishes, prepared_dishes)
plt.xlabel('Dishes')
plt.ylabel('Number of Dishes to Prepare')
plt.title('Optimal Number of Dishes to Prepare')
plt.xticks(rotation=45)
plt.show()
```
